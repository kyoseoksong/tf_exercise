# 초간단 Linear regression

### 그래프 정의하기

실제 데이터 

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

가중치와 바이어스를 변수로 정의

```python
weights = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
bias = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
```

실제 데이터를 feed해 주기 위한 placeholder를 정의

```python
x_true = tf.placeholder(tf.float32, name='x_true')
y_true = tf.placeholder(tf.float32, name='y_true')
```

Prediction 정의

```python
y_pred = tf.add(tf.multiply(weights, x_true), bias)
```

cost는 모든 (예측값 - 실제값) 을 제곱해 준 다음에 이를 평균을 낸 것으로 정의한다. 

```python
cost = tf.reduce_mean(tf.square(y_pred - y_true))
```

optimizer는 lr=0.1인 것으로 정의하고 cost를 최소화하는 train_op을 정의한다.

```python
optimizer = tf.train.AdamOptimizer(learning_rate=0.1)
train_op = optimizer.minimize(cost)
```

### 그래프 실행하기

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in xrange(300):
        _, cost_val = sess.run([train_op,cost], {x_true:x, y_true:y})
        print step, cost_val, sess.run(weights), sess.run(bias)
    print sess.run(y_pred, {x_true:5})
    print sess.run(y_pred, {x_true:10})
```
