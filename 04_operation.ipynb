# Operations

tf 1.0 이 되면서 사칙연산에 변경 생김
* tf.add() : 그대로
* tf.sub() : tf.subtract()
* tf.mul() : tf.multiply()
* tf.div() : 그대로 써도 되지만 가급적 tf.divide() 사용 권장됨

참고로 이들은 모두 element-wise 연산임

```python
tf.InteractiveSession()
```

```python
a = tf.constant([3,6])
b = tf.constant([2,2])

print tf.add(a,b).eval()
print tf.add_n([a, b, b]).eval() # a + b + b

print tf.multiply(a,b).eval()
# print tf.matmul(a, b) # error
print tf.matmul(tf.reshape(a, [1,2]), tf.reshape(b, [2,1])).eval()

tf.divide(a, b).eval()
tf.mod(a, b).eval()
```
